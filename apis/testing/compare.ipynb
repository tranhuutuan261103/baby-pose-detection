```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def compare_time(algorithms: list):
    # Initialize an empty list to hold the DataFrames for each algorithm
    all_dataframes = []

    # Loop over each algorithm and load the corresponding data
    for algorithm in algorithms:
        data = np.loadtxt(f'{algorithm}.txt', dtype=float)
        df = pd.DataFrame(data, columns=['time'])
        df['algorithm'] = algorithm  # Add a column to identify the algorithm
        all_dataframes.append(df)  # Add the DataFrame to the list

        # Calculate mean and standard deviation for the algorithm
        avg = np.mean(data)
        std = np.std(data)
        print(f'Trung bình (Mean) của {algorithm} keypoints: {avg * 1000:.4f} ms')
        print(f'Độ lệch chuẩn (Std Dev) của {algorithm} keypoints: {std * 1000:.4f} ms\n')

    # Combine all DataFrames into one for comparison
    combined_df = pd.concat(all_dataframes)

    # Create a violin plot for comparison
    plt.figure(figsize=(len(algorithms) * 4, 4))
    sns.violinplot(x='algorithm', y='time', data=combined_df, width=0.5, hue='algorithm', palette='Set1')

    # Add titles and labels
    plt.title('Comparison of Detection Time', fontsize=14)
    plt.xlabel('Keypoints', fontsize=12)
    plt.ylabel('Time (seconds)', fontsize=12)
    # i want y-axis to start from 0
    # plt.ylim(0, np.max(combined_df['time']) + 0.0001)

    # Display the plot
    plt.show()
```

```python
compare_time(["33", "15"])
```
